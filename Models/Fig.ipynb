```python
import pandas as pdb
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from os import listdir
from os.path import isfile, join
import re
import math
import pandas as pd
```

```python
def get_files_names():

    filenames = listdir()
    return [filename for filename in filenames if filename.endswith(".csv")]

def open_csv(files):
    
    tab = {}
    for file in files:
        tab[file] = pd.read_csv(file, index_col=0)
    return tab

def transformation(df):
    
    df = df.reset_index()
    df = pd.pivot_table(df, columns="index")
    df = df.reset_index()
    df = df.rename(columns={"index":"Np"})
    df["Np"] = pd.to_numeric(df["Np"])
    df = df.sort_values(by=["Np"])
    df = df.reset_index(drop=True)
    return df
```

```python
M1S1 = pd.read_csv("M1S1R10VTO90VTI25PT50.csv", index_col=0)
M2S1 = pd.read_csv("M2S1R10TP1VTO90VTI25VD70PT50PD3.3K1.csv", index_col=0)
M3S1 = pd.read_csv("M3S1R10TP1VTO90VTI25VD70PT100PD3.3K1.csv", index_col=0)
M1S1 = transformation(M1S1)[4:]
M2S1 = transformation(M2S1)[4:]
M3S1 = transformation(M3S1)[4:]

M1S2 = pd.read_csv("M1S2R10VTO90VTI25PT50.csv", index_col=0)
M2S2 = pd.read_csv("M2S2R10TP1VTO90VTI25VD70PT50PD3.3K1.csv", index_col=0)
M3S2 = pd.read_csv("M3S2R10TP1VTO90VTI25VD70PT100PD3.3K1.csv", index_col=0)
M1S2 = transformation(M1S2)[4:]
M2S2 = transformation(M2S2)[4:]
M3S2 = transformation(M3S2)[4:]

M1S3 = pd.read_csv("M1S3R10VTO90VTI25PT50.csv", index_col=0)
M2S3 = pd.read_csv("M2S3R10TP1VTO90VTI25VD70PT50PD3.3K1.csv", index_col=0)
M3S3 = pd.read_csv("M3S3R10TP1VTO90VTI25VD70PT100PD3.3K1.csv", index_col=0)
M1S3 = transformation(M1S3)[4:]
M2S3 = transformation(M2S3)[4:]
M3S3 = transformation(M3S3)[4:]
```

```python
fig = plt.figure(figsize=(15, 10))

plt.plot("Np", "Ct", data=M1S1,  color="green", linestyle="-", label="Model 1, Qdel = 1")
plt.plot("Np", "Ct", data=M1S2,  color="green", linestyle="--", label="     -       , Qdel = 2")
plt.plot("Np", "Ct", data=M1S3,  color="green", linestyle=":", label="     -       , Qdel = 3")

plt.plot("Np", "Ct", data=M2S1,  color="red", linestyle="-", label="Model 2, Qdel = 1")
plt.plot("Np", "Ct", data=M2S2,  color="red", linestyle="--", label="     -       , Qdel = 2")
plt.plot("Np", "Ct", data=M2S3,  color="red", linestyle=":", label="     -       , Qdel = 3")

plt.plot("Np", "Ct", data=M3S1,  color="orange", linestyle="-", label="Model 3, Qdel = 1")
plt.plot("Np", "Ct", data=M3S2,  color="orange", linestyle="--", label="     -       , Qdel = 2")
plt.plot("Np", "Ct", data=M3S3,  color="orange", linestyle=":", label="     -       , Qdel = 3")

plt.xticks([5, 10, 15, 20, 25, 30, 35, 40], fontsize=16)
plt.xlabel("Number of delivery points", fontsize=18)
plt.ylabel("Total cost ($/hour)", fontsize=18)
plt.legend(fontsize=16)
plt.grid()
plt.show()
plt.draw()
fig.savefig("Model100Ct")
```
